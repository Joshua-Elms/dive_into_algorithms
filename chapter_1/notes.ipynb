## Galilean Equations

* Position is modeled by his two? equations, I'm sure you'll find them if you go looking... 

```python
def ball_trajectory(x, v1 = 0.99, v2 = 9.9):
    """
    height of thrown ball as a function of horizontal position x, horizontal velocity v1, and vertical velocity v2
    """
    coef1 = v2 / v1
    coef2 = 9.81 / 2 * (v1**2)
    
    return (coef1 * x) - (coef2 * (x ** 2))

xs = [x/100 for x in list(range(209))]
ys = [ball_trajectory(x) for x in xs]
plt.plot(xs,ys)
plt.title('The Trajectory of a Thrown Ball')
plt.xlabel('Horizontal Position of Ball')
plt.ylabel('Vertical Position of Ball')
# plt.axhline(y = 0)
xs2 = [0.1,2.09]
ys2 = [ball_trajectory(0.1),0]
xs3 = [0.2,2.09]
ys3 = [ball_trajectory(0.2),0]
xs4 = [0.3,2.09]
ys4 = [ball_trajectory(0.3),0]
# plt.title('The Trajectory of a Thrown Ball - with Lines of Sight')
# plt.xlabel('Horizontal Position of Ball')
# plt.ylabel('Vertical Position of Ball')
# plt.plot(xs,ys,xs2,ys2,xs3,ys3,xs4,ys4)
xs5 = [0.3,0.3]
ys5 = [0,ball_trajectory(0.3)]
xs6 = [0.3,2.09]
ys6 = [0,0]
plt.title('The Trajectory of a Thrown Ball - Tangent Calculation')
plt.xlabel('Horizontal Position of Ball')
plt.ylabel('Vertical Position of Ball')
plt.plot(xs,ys,xs4,ys4,xs5,ys5,xs6,ys6)
plt.text(0.31,ball_trajectory(0.3)/2,'A',fontsize = 16)
plt.text((0.3 + 2)/2,0.05,'B',fontsize = 16)
plt.savefig("trajectory.png", dpi = 400)
```

The above figure is intended to demonstrate a cool fact about a theoretical individual catching a baseball; assuming that they are standing in the position the ball will land, the tangent of the angle formed by them looking at the ball in flight grows linearly w/ a coefficient of something like 5 given our constraints. 

In other words, an outfielder would know they are standing where the ball will land if they are forced to follow the ball by tipping their head back at a constant speed. 

This runs counter to the so called "analytical" approach of calculating the position a ball will land (determine distance and velocity, plug into position equations, etc), and is intended to demonstrate what might be a more human approach.
